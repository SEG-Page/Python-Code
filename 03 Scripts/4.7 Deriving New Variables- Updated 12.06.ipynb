#### Using IF functions

```python
# Define a function called price_label

def price_label(row):
    
  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Run a frequency check
df['price_range'].value_counts(dropna = False)
```

```python
# Check for the most expensive products
df['prices'].max()
```

#### Question 2 - Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method

```python
# Calculate the two most busiest days and the two slowest days
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value <= 1: # 0, 1 Saturday and Sunday 
    result.append("Busiest days")
  elif value >= 3 and value <=4: # 3, 4 Tuesday and Wednesday
    result.append("Slowest days")
  else: #2, 5, 6, Monday, Thursday, Friday
    result.append("Regularly busy")
```

```python
#Create a column with Busiest Days
df_ords_prods_merged['busiest_days'] = result
```

#### Question 4 - When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Check output for merged data
df_ords_prods_merged.head()
```

```python
# Run a frequency check for orders_day_of_the_week
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Finding and grouping the busiest hour of the day based on frequency check above

# Count orders in millions: hours 10, 11, 14, 15, 13, 12, 16, 9, 17, 8, 18, 19 as Most Orders
# Count orders in hundreds of thousands: hours 20, 7, 21, 22, 23, 6, 0, 1 as Average Orders
# Count orders in tens of thousands: hours 5, 2, 4, 3 as Fewest Orders
```

```python
# Use loc() function to group busiest hour - Most Orders
df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'].isin([10, 11, 14, 15, 13, 12, 16, 9, 17, 8, 18, 19]),'busiest_period_of_day')] = 'Most Orders'
```

```python
# Use loc() function to group busiest hour - Average Orders
df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'].isin([20, 7, 21, 22, 23, 6, 0, 1]),'busiest_period_of_day')] = 'Average Orders'
```

```python
# Use loc() function to group busiest hour - Fewest Orders
df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'].isin([5, 2, 4, 3]),'busiest_period_of_day')] = 'Fewest Orders'
```

```python
# Run a frequency check for new column busiest period of day
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
